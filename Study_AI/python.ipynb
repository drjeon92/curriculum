```python
# 네이버 영화랭킹 크롤링

# 1. requests, BeautifulSoup 모듈 불러오기
import requests
from bs4 import BeautifulSoup
```

```python
# url 지정
url = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn'

# requests get 통신
res = requests.get(url)

# 응답코드 : 200
# print(res.status_code)

# BeautifulSoup 객체 생성, 네이버 영화 사이트 html 파싱
soup = BeautifulSoup(res.text, 'html.parser')
# print(soup.prettify())

tit3_text = soup.select(".tit3 > a")

for text in tit3_text:
    # print(text.string)
    with open('move_rank.txt', 'a') as f:
        f.write(text.string+'\n')
```
