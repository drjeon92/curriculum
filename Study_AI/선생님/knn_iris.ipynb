```python
data = pd.DataFrame(iris_dataset["data"], columns=iris_dataset["feature_names"])
data
```

```python
data["target"] = iris_dataset["target"]
```

```python
data
```

```python
data = data.set_index("target", drop=False)
data
```

```python
plt.scatter(data.loc[0, "petal length (cm)"], data.loc[0, "petal width (cm)"])
plt.scatter(data.loc[1, "petal length (cm)"], data.loc[1, "petal width (cm)"])
plt.scatter(data.loc[2, "petal length (cm)"], data.loc[2, "petal width (cm)"])
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
```

```python
pd.plotting.scatter_matrix(data.iloc[:, :-1],c=data.loc[:, "target"],
                           figsize=(20, 20), marker='o',
                           hist_kwds={'bins': 20}, s=60, alpha=.8)
```

```python
data.boxplot(figsize=(10,8))
```
