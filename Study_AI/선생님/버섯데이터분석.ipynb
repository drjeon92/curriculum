```python
# 원핫인코딩 시작
X.loc[:, "cap-shape"].unique()
```

```python
cap_shape_d = pd.get_dummies(X.loc[:, "cap-shape"])
```

```python
X = pd.concat([X, cap_shape_d], axis=1)
X
```

```python
X.drop(['b', 'c', 'f','s','x', 'k'], axis=1, inplace=True)
```

```python
X
```

```python
X_one = pd.get_dummies(X, prefix=X.columns)
```

```python
X_one
```

```python
X_train, X_test, y_train, y_test= train_test_split(X_one, y, test_size=0.2)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```
