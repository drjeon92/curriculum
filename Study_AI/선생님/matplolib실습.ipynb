```python
import matplotlib.pyplot as plt
```

```python
data = [1, 2, 3, 4]
plt.plot(data)
```

```python
data_y = [1, 3, 10, 50]
plt.plot(data, data_y, '*')
plt.xlabel("X")
plt.ylabel("y")
#plt.axis([0, 10, 0, 100])      # xmin, xmax, ymin, ymax
```

```python
data_x = [1, 4, 9, 16]
plt.figure(figsize=(20, 10))
plt.plot(data_x, [1, 2, 3, 4], 'r', marker="*", linewidth=10)
plt.plot(data_x, [1, 4, 8, 16], 'b', marker="^", markersize=10)
plt.plot(data_x, [3, 6, 9, 12], 'g')
plt.grid(True, axis='x')
```

```python
data = pd.read_csv("data/bmi.csv", index_col="label")
```

```python
data
```

```python
plt.bar("fat", data.loc["fat"].count())
plt.bar("normal", data.loc["normal"].count())
plt.bar("thin", data.loc["thin"].count())
```

```python
plt.bar("fat", data[data["height"] >= 170].loc["fat"].count())
plt.bar("normal", data[data["height"] >= 170].loc["normal"].count())
plt.bar("thin", data[data["height"] >= 170].loc["thin"].count())
```
