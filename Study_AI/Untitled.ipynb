```python
import matplotlib.pyplot as plt
```

```python
data = [1,2,3,4]
plt.plot(data)
```

```python
data_y = [1, 3, 10, 50]
plt.plot(data,data_y, "*")
plt.xlabel("x")
plt.ylabel("y")
# plt.axis([0,10, 0,100]) # (xMin, xMax, ymin, yMax)
```

```python
data_x = [1,4,9,16]
plt.figure(figsize=(20,10))
plt.plot(data_x, [1,2,3,4], 'r', marker="*", linewidth=10)
plt.plot(data_x, [1,4,8,16], 'b', marker="^", markersize=10)
plt.plot(data_x, [3,6,9,12], 'g' )
plt.grid(True, axis="y")
```
